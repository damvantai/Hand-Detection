```python
import scipy.io as sio
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
```

```python
path = "/home/damvantai/Documents/datasets/hand_dataset/egohands/_LABELLED_SAMPLES/CARDS_COURTYARD_H_S/"

image_path_array = []

# All path image in image_path_array
for i in os.listdir(path):
    if (i.split(".")[-1] == "jpg"):
        temp_path = path + i
        image_path_array.append(temp_path)
```

```python
image_path_array.sort()
```

```python
# Read mat file
boxes = sio.loadmat(path + "polygons.mat")
polygons = boxes["polygons"][0]

point_index = 0
```

```python
# In polygons is folder image, first is a image
all_row = ""
for first in polygons:
    row = ""
    index = 0
    
    # image path 
    img_path = image_path_array[point_index]
    
    row += img_path
    point_index += 1
    boxarray = []
    csvholder = []
    
    for pointlist in first:
        pst = np.empty((0, 2), int)
        max_x = max_y = min_x = min_y = height = width = 0
        
        findex = 0
        for point in pointlist:
            if (len(point) == 2):
                x = int(point[0])
                y = int(point[1])
                
                if (findex == 0):
                    min_x = x
                    min_y = y
                findex += 1
                max_x = x if (x > max_x) else max_x
                min_x = x if (x < min_x) else min_x
                max_y = y if (y > max_y) else max_y
                min_y = y if (y < min_y) else min_y
                # print(index, "====", len(point))
                appeno = np.array([[x, y]])
                pst = np.append(pst, appeno, axis=0)
        
        row += " " + str(min_x) + "," + str(max_y) + "," + str(max_x) + "," + str(min_y) + "," + str(0)
        
    all_row += row + "\n"
```

```python
text_file = open("row_egohands.txt", "w")

text_file.write(all_row)

text_file.close()
```
