```python
import os
import glob
import matplotlib.pyplot as plt
import cv2
import numpy as np
import dlib
import scipy.io as sio
```

```python
image_path = "/home/damvantai/Documents/datasets/hand_dataset/hand_dataset/training_dataset/training_data/images/"
```

```python
mat_path = "/home/damvantai/Documents/datasets/hand_dataset/hand_dataset/training_dataset/training_data/annotations/"
```

```python
def convert_xml_image(file_image):
#     print(file_image)
    # Split name file
    file_mat = file_image.split('.')[0] + ".mat"
    
    # Full path mat and image
    full_path_image = image_path + file_image
    full_path_mat = mat_path + file_mat

    # Load image and mat
    image = cv2.imread(full_path_image)

    data = sio.loadmat(full_path_mat)
    
    # Number hand in a image
    number_hand = data['boxes'][0].shape[0]
    
    hands_image = []
    xml_image = ""
    
    # Data a hand (top, left, width, height)
    hand_data = np.empty((4))
    
    # Data a hand in file mat
    hand_position = np.empty((4, 2))
    
    xml_image += "<image file='" + str(file_image) + "'>\n"
#     print(xml_image)
    for i in range(0, number_hand):
        # Data a hand in file mat convert array
        for j in range(0, 4):
            hand_position[j] = data['boxes'][0][i][0][0][j]
#             print(hand_position[j])
        
        # Data a hand (top, left, width, height)
        top_left = [int(hand_position.transpose()[0].min()), int(hand_position.transpose()[1].min())]
        bottom_right = [int(hand_position.transpose()[0].max()), int(hand_position.transpose()[1].max())]
        width = bottom_right[0] - top_left[0]
        height = bottom_right[1] - top_left[1]
        
        hand_data = [top_left[0], top_left[1], width, height]
        hands_image.append(hand_data)
        

    for i in hands_image:
        xml_image += "\t<box top='" + str(i[0]) + "' left='" + str(i[1]) + "' width='" + str(i[2]) + "' height='" + str(i[3]) + "'/>\n"
    xml_image += "</image>\n"
        
    return xml_image
```

```python
print(convert_xml_image("VOC2010_711.jpg"))
```

```python
xml_all = ""
for file_image in os.listdir(image_path):
#     print(file_image)
    xml_all += convert_xml_image(file_image)
```
